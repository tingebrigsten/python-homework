```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import style
style.use("ggplot")
```

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=40, centers=2, random_state=42, cluster_std=1.25)
plt.scatter(X[:, 0], X[:, 1], c=y, s=100, cmap="bwr")
plt.show()
```

```python
# Support vector machine linear classifier
# rbf: creates a non-linear hyperplane
# linear: creates a linear, 2D hyperplane
# poly: creates a non-linear hyperplane

from sklearn.svm import SVC
model = SVC(kernel='linear')
model.fit(X, y)
```

```python
# Plot the decision boundaries
x_min = X[:, 0].min()
x_max = X[:, 0].max()
y_min = X[:, 1].min()
y_max = X[:, 1].max()
print(x_min, x_max, y_min, y_max)
```

```python
# Store decision boundaries in mesh grid
XX, YY = np.mgrid[x_min:x_max, y_min:y_max]
print(XX)
```

```python
# Use the decision_function function to identify sides of the hyperplane
# np.c_ is used to concat the data into one array.
# .ravel() is used to flatten a 2D array to a 1D array.
Z = model.decision_function(np.c_[XX.ravel(), YY.ravel()])
print(Z)
```

```python
# Put the result into a color plot

Z = Z.reshape(XX.shape) # reshape result into the shape of XX. 
# plt.pcolormesh(XX, YY, Z > 0, cmap=plt.cm.Paired)
plt.contour(XX, YY, Z, colors=['k', 'g', 'k'],
            linestyles=['--', '-', '--'], levels=[-1, 0, 1])
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', edgecolor='k', s=100)
plt.show()
```

```python
# Demonstrate that support vectors will sometimes cross the margin
X, y = make_blobs(n_samples=100, centers=2, random_state=0, cluster_std=.95)
plt.scatter(X[:, 0], X[:, 1], c=y, s=100, cmap="bwr")
plt.show()
```

```python
# Split data into training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Fit to the training data and validate with the test data
model = SVC(kernel='linear')
model.fit(X_train, y_train)
predictions = model.predict(X_test)
predictions
```

```python
# Plot the decision boundaries
x_min = X[:, 0].min()
x_max = X[:, 0].max()
y_min = X[:, 1].min()
y_max = X[:, 1].max()

XX, YY = np.mgrid[x_min:x_max, y_min:y_max]
Z = model.decision_function(np.c_[XX.ravel(), YY.ravel()])

# Put the result into a color plot
Z = Z.reshape(XX.shape)
# plt.pcolormesh(XX, YY, Z > 0, cmap=plt.cm.Paired)
plt.contour(XX, YY, Z, colors=['k', 'g', 'k'],
            linestyles=['--', '-', '--'], levels=[-1, 0, 1])
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', edgecolor='k', s=100)
plt.show()
```

```python
# Calculate classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions,
                            target_names=["blue", "red"]))
```
